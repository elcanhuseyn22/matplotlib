```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
y = np.arange(2,11,2)

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.2,0.3])

#axes1-------
axes1.plot(y,x,"red")

axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

#axes2-------
axes2.plot(x,y,"red")

axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")



plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1)

axes[0].plot(x,y,"red")
axes[0].set_title("First axes")
axes[1].plot(x,x**4,"red")
axes[1].set_title("Second axes")
plt.tight_layout()
```

### Fiqur özəlləşdirmə

```python
fig = plt.figure(figsize = (3,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,"red")

plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize = (8,6))

axes[0].plot(x,y,"red")
axes[0].set_title("First Axes")
axes[1].plot(x,x**2,"purple")
axes[1].set_title("Second Axes")
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (6,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**0.5,color = "red",label = "X kökaltı")

axes.plot(x,x**2,color = "yellow",label = "X kvadrat")

axes.plot(x,x**3,color = "#52FF33",label = "X kub")

axes.legend()
plt.show()
```

### Qrafik özəlləşdirmə

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "red",linewidth = 3,linestyle = "-.",marker = "o",markersize = 8)

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "red",linewidth = 3,linestyle = "-.",marker = "o",markersize = 8)

axes.set_xlim(0,10)
axes.set_ylim(0,40)
plt.show()
```
