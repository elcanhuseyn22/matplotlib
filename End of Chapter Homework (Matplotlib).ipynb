```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(1,51)
y = np.arange(1,101,2)
z = x ** 3
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color ="red")

axes.set_xlabel("X values")
axes.set_ylabel("Y values")
axes.set_title("X vs Y Graph")


plt.show()
```

```python
fig = plt.figure(figsize = (8,6))

axes1 = fig.add_axes([0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

#axes1
axes1.plot(x,z,color = "purple")
axes1.set_title("Outer Graph - X vs Z Graph")
#axes2
axes2.plot(x,y,color = "red")
axes2.set_title("Inner Graph - X vs Y Graph")
```

```python
fig,axes = plt.subplots(nrows =1 ,ncols = 2,figsize=(10,6))

axes[0].plot(x,y,color = "red",linewidth = 3,linestyle = "--",marker = "o",markerfacecolor = "black")
axes[0].set_xlabel("X values")
axes[0].set_ylabel("Y values")
axes[0].set_title("X vs Y Graph")


axes[1].plot(x,z,color = "purple",linewidth = 3,linestyle = "-.")
axes[1].set_xlabel("X values")
axes[1].set_ylabel("Z values")
axes[1].set_title("X vs Z Graph")

plt.tight_layout()
plt.show()
```
